# 모델 로딩

```python
import torch
from transformers import pipeline
```

Transformers는 허깅페이스에서 제공하는 라이브러리

```python
torch.cuda.is_available()
```

GPU 사용할 수 있냐? > [런타임] > 런타임 유형 > GPU로 설정되어있으면 됨

```python
model_id = "meta-llama/Llama-3.2-3B-Instruct"
```

hubingface에서 가져온 모델이름을 붙여넣어준다

뒤에 내용을 보고 앞에 나올 말을 예상해서 한다 > Instruct > 다음에 나올말은 확률이 높은걸 내뱉는다
학습을 시키는데 이러한 문장들에 어떤 대답이 나올것을 매핑시켜서 학습 시킨 것이다.

```python
pipe = pipeline(
    "text-generation",
    model=model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto",
)
```

pytorch 타입은 16 타입
모델 다운로드 받는 것


```python
messages = [
    {"role": "system", "content": "You are a pirate chatbot who always responds in pirate speak!"},
    {"role": "user", "content": "Who are you?"},
]
```

content는 해적처럼 대답하는 챗봇이다, content는 너누구니? 이렇게 질문해라

```python
outputs = pipe(
    messages,
    max_new_tokens=256,
)
```

답의 토큰은 265자로 해라

```python
print(outputs[0]["generated_text"][-1]["content"])
```
